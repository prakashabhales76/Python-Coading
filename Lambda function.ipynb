```python
# convert a list of integers to theire corisponding square values:  
# map in the maping the list every one purticuler function to in of the lambda function
# list:list the all numbers
numbers=[1,2,3,4,5,6]
squares=list(map(lambda x:x**2,numbers))
print(squares)
```

```python
## filter out even number from a list:# lambada use filtter of maping
numbers=[1,2,3,4,5,6]
list(filter(lambda x:x%2==0,numbers)) # even number
```

```python
people=[
    {'name':'Alice','age':25,'occupantion':'Engineer'},
    {'name':'Bob','age':30,'occupantion':'Manager'},
    {'name':'Chaelie','age':22,'occupantion':'Intern'},
    {'name':'Dave','age':27,'occupantion':'Designer'},
]
```

```python
sorted(people,key=lambda x:x['age']) # age wise sorted   # sorting with lambda
```

```python
sorted(people,key=lambda x:(x['age'],x['name'])) #2 key wise age and key
```

```python
from itertools import groupby
```

```python
words=['apple','banana','cherry','date','elderbarry','fig']     # list
groups=groupby(sorted(words),key=lambda x: x[0])
for key, group in groups:
    print(key,list(group))
```
