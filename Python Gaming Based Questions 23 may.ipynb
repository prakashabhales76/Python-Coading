### Question 1:
Create a math quiz game where the player is presented with a series of addition questions. For each correct answer, the player earns 5 points, and for each incorrect answer, deduct 2 points. Play the game for 10 rounds and calculate the total score. If the total score is positive, declare the player as the winner.

### Question 2. Solve the following game.

Step 1. Generate a random number between 1 to 100. 

Step 2. Accept another number from user in the same range. 

Step 3. Generate score of the round : 

Rule 1: If both numbers have 1 factor common - Score 1

Rule 2: If both numbers have 2 factors common - Score 2

Rule 3: If both numbers have 3 factors common - Score 3

Rule 4: If both numbers have more than 3 factors common - Score 5

Generate 3 rounds of game. Cumulate the scores of each round. If total score is 10 or more. Declare winner.    

### Question 3:
Create a word guessing game where the player has to guess a randomly chosen word from a list. The player can guess one letter at a time. For each correct guess, the player earns 3 points. If the guess is incorrect, deduct 1 point. Play the game for 3 rounds and calculate the total score. If the total score is equal to or higher than 15, declare the player as the winner.

### Question 4:
Create a dice rolling game where the player rolls two dice. If the sum of the dice is even, the player earns 2 points. 
If the sum is odd, the player loses 1 point. Play the game for 5 rounds and calculate the total score. 
If the total score is positive, declare the player as the winner.
